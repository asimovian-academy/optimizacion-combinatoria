```python
def SingletonsGen(B):
    lista = []
    for b in B:
        lista.append([b])
    return lista

def ImagesGen(B, N, test=False):
    k=1; l=len(B)
    X = SingletonsGen(B)#; print "X=",X    
    #X=[]
    #for i in range(l**N):
    #    X.append([])
    Y = []
    while k < N:
        for x in X:
            for b in B:
                y=copy(x)#; print x
                y.append(b)#; print y
                Y.append(y)
        X=copy(Y)
        Y=[]
        k+=1
        if test: print "ImageGen: step "+str(k)+" X="+str(X)
    return X
```

```python
def FunGen(A,B, test=False):
    N=len(A)
    Ims=ImagesGen(B,N, test)
    Funs = []
    for im in Ims:
        fun = []
        for k,a in enumerate(A):
            fun.append([a,im[k]])
        if test: print "fun= "+str(fun)
        Funs.append(fun)
    return Funs
A=range(3); B = range(3); test=False
Funs = FunGen(A,B, test)
```

```python
def onetoone(fun):
    for k, pair in enumerate(fun):
        x=pair[0]
        y=pair[1]
        for i in range(k):
            xx = fun[i][0]
            yy = fun[i][1]
            if y==yy:
                return False
    return True

for fun in Funs:
    if onetoone(fun):
        #print fun            
        True
```

```python
def image(fun):
    lista=[]
    for pair in fun:
        y = pair[1]
        lista.append(y)
    return list(Set(lista))

for fun in Funs:
    #print fun, image(fun)
    True
```

```python
def onto(fun, B):
    if Set(image(fun))==Set(B):
        return True
    else:
        return False

for fun in Funs:
    if onto(fun, B):
        #print fun, image(fun)
        True

def invertible(fun, B):
    if onetoone(fun) and onto(fun, B):
        return True
    else:
        return False
```

```python
def FunAnalyser(A,B):
    Funs = FunGen(A, B)
    mitabla=[["Function", "1:1", "Onto", "Invertible"]]
    for fun in Funs:
        renglon=[fun, onetoone(fun), onto(fun,B), invertible(fun, B)]
        mitabla.append(renglon)
    return mitabla

A=range(3); B = range(4)

show(table(FunAnalyser(A,B)))
```
